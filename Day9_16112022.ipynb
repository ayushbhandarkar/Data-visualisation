```python
labels = ['bat', 'cat', 'rat', 'dog']
sizes = [15,25,60,100]
```

```python
# ax = plt.subplot(1,2, figsize = (12,4))
fig = plt.subplots(figsize = (5,5),  labels = labels, dpi = 100, autopct = '%%\f%%', shadow = True, startangle = 00)
# explode = (0.1,0,0)
# patches, text, autotextx = ax.pie()
# plt.setp(autotexts, size = 10, color = 'white')

plt.pie(sizes)
plt.show()
```

```python
import numpy as np


explode = (0.1,0.0,0.1, 0.0)
colors = ( "orange", "cyan", "brown", "grey",)

# Wedge properties
wp = { 'linewidth' : 1, 'edgecolor' : "green" }

def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:.1f}%\n({:d} g)".format(pct, absolute)

# Creating plot
fig, ax = plt.subplots(figsize =(10, 7))
wedges, texts, autotexts = ax.pie(sizes,
                                  autopct = lambda pct: func(pct, sizes),
                                  explode = explode,
                                  labels = labels,
                                  shadow = True,
                                  colors = colors,
                                  startangle = 90,
                                  wedgeprops = wp,
                                  textprops = dict(color ="magenta"))

# Adding legend
ax.legend(wedges, labels,
          title ="Labels",
          loc ="center left",
          bbox_to_anchor =(1, 0, 0.5, 1))


plt.setp(autotexts, size = 8, weight ="bold")
ax.set_title("Customizing pie chart")
 
# show plot
plt.show()
```

```python
import pandas as pd

db = pd.read_csv('titanic.csv')
db.head()
```

```python
# explode = (0.1,0.0,0.1, 0.0)
colors = ( "orange", "cyan", "brown", "grey",)

# Wedge properties
wp = { 'linewidth' : 1, 'edgecolor' : "green" }

def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:.1f}%\n({:d} g)".format(pct, absolute)

# Creating plot
fig, ax = plt.subplots(figsize =(10, 7))
wedges, texts, autotexts = ax.pie(db['Survived'],
                                  autopct = lambda pct: func(pct, sizes),
                                  labels = db['Sex'],
                                  shadow = True,
                                  colors = colors,
                                  startangle = 90,
                                  wedgeprops = wp,
                                  textprops = dict(color ="magenta"))

# Adding legend
ax.legend(wedges, db['Sex'],
          title ="Labels",
          loc ="center left",
          bbox_to_anchor =(1, 0, 0.5, 1))


plt.setp(autotexts, size = 8, weight ="bold")
ax.set_title("Customizing pie chart")
 
# show plot
plt.show()
```
